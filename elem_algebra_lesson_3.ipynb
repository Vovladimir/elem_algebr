### 3. Задание

1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_cart(a, r=1):
    if r < 0:
        print (f'{r} радиус должен быть больше или равен нулю!')
    elif 0 > a > 360:
        print (f'{a} угол должен быть в пределах от 0 до 360 градусов!')
    else:
        x = round(r * np.cos(a * (np.pi / 180)), 2)
        y = round(r * np.sin(a * (np.pi / 180)), 2)
        return x,y

polar_to_cart(270)
polar_to_cart(120, 5)
```

### 4. Задание (в программе):
Решите систему уравнений:

1. y = x2 – 1

   exp(x) + x(1 – y) = 1


2. y = x2 – 1

   exp(x) + x∙(1 – y) > 1


1. Решение:

y = x2 - 1

y = (exp(x) + x - 1) / x

```python
x = np.linspace(-np.pi, np.pi, 201)

plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1) / x)
xlabel('x')
ylabel('y')
plt.ylim(-5,8) 
plt.grid(True)


def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (3, 6))
x2, y2 =  fsolve(equations, (-1, 2))
              
print (f'x1 = {round(x1, 2)}, y1 = {round(y1, 2)}')
print (f'x2 = {round(x2, 2)}, y2 = {round(y2, 2)}')
```
